```python
import requests
from bs4 import BeautifulSoup
```

```python
url = "https://www.hastane.com.tr/hastaneler.html"
```

```python
html = requests.get(url).content
```

```python
soup = BeautifulSoup(html, "html.parser")
```

```python
il_listesi = soup.find("table", { "id" : "ContentPlaceHolder1_dlCitiesList" })
```

```python
links = []
for i in il_listesi:
    for j in i:
        try:
            links.append("https://www.hastane.com.tr"+j.find('a').get('href'))
            
        except:
            pass
```

```python
len(links)
```

```python
links[:5]
```

```python
# tüm illerin listelerini aldık. Şimd tek bir ilin hastanele listesini alalım. Daha sonra generik hale getirelim.
```

```python

html = requests.get(links[0]).content
soup = BeautifulSoup(html,'html.parser',from_encoding="utf8")
hastane_listesi = soup.find("table", { "id" : "ContentPlaceHolder1_dlHospitals" })
```

```python
hastane_isimleri = []
for i in hastane_listesi:
    for j in i:
        try:
            hastane_isimleri.append(j.find('a',{'class':'MainLink'}).text)
            
        except:
            pass
```

```python
hastane_isimleri
```

```python
# bazı illerin ekstra sayfaları var.
sayfa_linkleri = []
for link in links:
    for sayfa in range(2,10):
        href = link.split('.html')
        sayfa_linkleri.append(''.join(href)+f'-sayfa-{sayfa}.html')
```

```python
link_tree = links + sayfa_linkleri
```

```python
len(link_tree)
```

```python
# tüm linklerde yer alan hastane bilgilerinin alınması
```

```python
hastane_isimleri = []
for link in link_tree:
    html = requests.get(link).content
    soup = BeautifulSoup(html,'html.parser',from_encoding="UTF-8")
    try:
        hastane_listesi = soup.find("table", { "id" : "ContentPlaceHolder1_dlHospitals" })
    except:
        continue
    try:
        for i in hastane_listesi:
            for j in i:
                try:
                    name = j.find('a',{'class':'MainLink'}).text
                    hastane_isimleri.append(name)
                    print(name)
                    
                except:
                    pass
    except:
        pass
```

```python
hastane_isimleri
```

```python
import pandas as pd
```

```python
hastane_df = pd.DataFrame({'hastane ismi':hastane_isimleri})
```

```python
hastane_df['hastane ismi'] = hastane_df['hastane ismi'].map(lambda x: x.replace('ý','ı').replace('ð','ğ').replace('Ý','İ').replace('Ð','Ğ').replace('þ','ş'))
```

```python
hastane_df
```

```python
hastane_df.to_csv('hastane.csv')
```
